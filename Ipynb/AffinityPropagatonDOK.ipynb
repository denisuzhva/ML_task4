```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
## Hyperparams and metadata

N_ITERS = 1
NODES = 196591
EDGES = 950327
```

```python
## Load data

adj_path = '../Dataset/adj_dok.npy'
s_dok = np.load(adj_path)
print(s_dok.shape)

# Visualize
plt.figure(figsize=(7, 5))
plt.hist2d(s_dok[:, 0], s_dok[:, 1], cmap='inferno', bins=300, norm=LogNorm())
plt.colorbar()
plt.xticks(np.arange(0, NODES, NODES // 5))
plt.yticks(np.arange(0, NODES, NODES // 5))
plt.show()
```

```python
## Set self-similarity

self_sim = 10
nodes_arange = np.arange(NODES).reshape([-1, 1])
self_vals = np.full((NODES, 1), self_sim, dtype=np.int32)
self_edges = np.concatenate((nodes_arange, nodes_arange, self_vals), axis=1)

s_dok = np.concatenate((s_dok, self_edges), axis=0)
```
