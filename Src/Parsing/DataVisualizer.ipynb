```python
import numpy as np
import scipy.sparse as sp
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
data_path = '../../Dataset/loc-gowalla_edges.txt'
gowalla_edges = np.loadtxt(data_path).T
print(gowalla_edges.shape)
```

```python
plt.figure(figsize=(12, 10))
plt.hist2d(gowalla_edges[0], gowalla_edges[1], bins=300, cmap='inferno', norm=LogNorm())
plt.colorbar()
plt.show()
```

```python
s_mat = sp.load_npz('../../Dataset/adj_mat.npz')
```

```python
plt.spy(s_mat, markersize=0.1)
plt.show()
```

```python
n_nodes = s_mat.shape[0]
min_shape = 100000
min_rows = []
for row in range(n_nodes):
    shape_val = sp.find(s_mat[row])[2].size
    if shape_val < min_shape:
        min_shape = shape_val
        min_rows.append(row)
```

```python
print(min_shape)
print(min_rows)
```
