```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import LogNorm
```

```python
data_path = '../../Dataset/loc-gowalla_edges.txt'
gowalla_edges = np.loadtxt(data_path).T
print(gowalla_edges.shape)
```

```python
plt.figure(figsize=(12, 10))
plt.hist2d(gowalla_edges[0], gowalla_edges[1], bins=300, cmap='inferno', norm=LogNorm())
plt.colorbar()
plt.show()
```
